```python
import numpy as np
```

```python
# Parameter 
h = 0.002       # m 
delP = 5        # Pa
mu = 8.90e-4    # Pa*s
L = 0.016       # m
W = 0.004       # m
```

```python
def V(y,z):
    
    result = 0 
    
    for n in range(0,100):
        
        if n % 2 == 1:
            body1 = 1/(n**3) *(1- (np.cosh(n*np.pi*y/h)/np.cosh(n*np.pi*W/2/h)))
            body2 = np.sin(n * np.pi* z / h)
            coeff = 4 * h**2 * delP / np.pi**3 / mu / L 
            result += coeff * body1 * body2
       
    return result

# let us make an contour plot 

y = np.linspace(-W/2, W/2, 10)
z = np.linspace(0, h, 10)

Y, Z = np.meshgrid(y, z)

%matplotlib inline
import matplotlib.pyplot as plt
plt.title('2D MODEL')
plt.plot(Y.flatten(), Z.flatten(), 'k.')
plt.xlabel('width [m]')
plt.ylabel('height [m]')
plt.show()

Velocity = V(Y,Z)
plt.title('Contour Plot')
plt.contour(Y, Z, Velocity, levels=np.linspace(0, 0.2, 100))
plt.colorbar()
plt.xlabel('width [m]')
plt.ylabel('height [m]')
plt.show()

Velo = V(0,h/2)
print(f'maxium velocity is {Velo:1.2f} m/s')
```
